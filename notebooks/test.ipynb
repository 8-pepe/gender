```python
class Deck:
    """
    This is the Deck Class. This object will create a deck of cards to initiate
    play. You can then use this Deck list of cards to split in half and give to
    the players. It will use SUITE and RANKS to create the deck. It should also
    have a method for splitting/cutting the deck in half and Shuffling the deck.
    """
    def __init__(self):
        self.cards = []
        for suit in SUITE:
            for rank in RANKS:
                self.cards.append(rank+suit)
        print("deck created")
    
    def shuffle(self):
        shuffle(self.cards)
        print("deck shuffled")
    
    def split_deck(self):
        
        while len(self.cards)>0:
            deal_card = self.cards.pop(0)
            p1.hand.cards.append(deal_card)
            deal_card = self.cards.pop(0)
            p2.hand.cards.append(deal_card)
```

```python
class Hand:
    '''
    This is the Hand class. Each player has a Hand, and can add or remove
    cards from that hand. There should be an add and remove card method here.
    '''
    def __init__(self):
        self.cards = []
    
    def __str__(self):
        return f"Contains {len(self.cards)} cards"
    
    def add_cards(self,cards):
        self.cards.extend(cards)
    
    def remove_card(self):
        return self.cards.pop(0)
        
    def remove_three_cards(self):
        return [self.cards.pop(0) for x in range(3) ]
        
        
```

```python
class Player:
    """
    This is the Player class, which takes in a name and an instance of a Hand
    class object. The Payer can then play cards and check if they still have cards.
    """
    def __init__(self,name):
        self.name = name
        self.hand = Hand()
    
    def play_card(self):
        return self.hand.remove_card()
    
```

```python
# Gameplay
game_on = True
pot_cards = []
round_num = 0

while game_on:
    round_num += 1 
    print(f"Round: {round_num}" )
    p1_card = p1.play_card()
    p2_card = p2.play_card()
    print(f"Player1 Card: {p1_card}")
    print(f"Player2 Card: {p2_card}")
    
    pot_cards.extend([p1_card, p2_card])

    if VALUE[p1_card[:-1]] > VALUE[p2_card[:-1]]:
        #print(f"p1: {p1_card} won angainst p2: {p2_card}")
        p1.hand.add_cards(pot_cards)
        pot_cards = []
    elif VALUE[p1_card[:-1]] < VALUE[p2_card[:-1]]:
        #print(f"p1: {p1_card} lost angainst p2: {p2_card}")
        p2.hand.add_cards(pot_cards)
        pot_cards = []
    elif VALUE[p1_card[:-1]] == VALUE[p2_card[:-1]]:
        print("War")
        if len(p1.hand.cards) > 3:
            cards_1 = p1.hand.remove_three_cards()
            pot_cards.extend(cards_1)
        else:
            print("Player 2 won")
            game_on = False
            break
        if len(p2.hand.cards) > 3:
            cards_2 = p2.hand.remove_three_cards()
            pot_cards.extend(cards_2)
        else:
            print("Player 1 won")
            game_on = False
            break
        
    game_on = check_win(game_on)
    

    
```

```python
def multi_re_find(patterns, phrase):
    for pat in patterns:
        print("Searching for pattern :", pat)
        print(re.findall(pat, phrase))
        print("\n")
```
